#### Converting date_time to date_time format and extracting year month day and hour

```python
df['Date_Time'] = pd.to_datetime(df['Date_Time'])
```

```python
df
```

```python
df
```

```python
df['year'] = df["Date_Time"].dt.year
df['month'] = df["Date_Time"].dt.month
df['day'] = df["Date_Time"].dt.day
df['hour'] = df["Date_Time"].dt.hour
```

#### Droping duplicates where the Location + date time are same

```python
df_duplicates = df[df.duplicated(subset = ['Location','Date_Time'], keep = False)]
df_duplicates
```

```python
df_duplicates[(df_duplicates["Location"] == "New York") & (df_duplicates["Date_Time"] == "2024-01-12 18:05:38")]
```

#### This cannot be possible as at same time and same location two temperatures and not possible so we will drop one of them

```python
df.drop_duplicates(subset=["Location","Date_Time"],inplace = True)
```

```python
df
```

#### Checking range for each column

#### Here all the columns are in range

```python
# Standardizing the names like New york , new york to New York
df['Location'] = df['Location'].str.strip().str.title() 
```

```python
df
```

```python
#### ADDING is_weekend column
```

```python
df['day_of_week'] = df["Date_Time"].dt.dayofweek   # 0=Monday, 6=Sunday
df['is_weekend'] = df['day_of_week'].isin([5,6]).astype(int)
```

```python
df
```

#### Accoring to month specifying season (Approx)

```python
def get_season(month):
    if month in [12,1,2]: return 'Winter'
    elif month in [3,4,5]: return 'Spring'
    elif month in [6,7,8]: return 'Summer'
    else: return 'Fall'

df['season'] = df['month'].apply(get_season)
```
